**Dashboard Implementation (The "Final Product")**  

*Network Visualization (The "Villa Sociogram")*
- Time Selector: Granular toggling (Weekly default, scalable to custom event-based windows).
- Nodes: Size-weighted by mention_share, color-coded by Louvain_cluster.
- Edges: Weighted by co-mention_count.
- Tooltips: Deep-dive metadata including LLM-distilled perspectives, toxicity scores, and approval/polarization metrics.

*Temporal Tracking (The "Contestant Arc")*
- Line-Chart Interface: Multi-line visualization tracking contestant-specific metrics over time.
- Dynamic Metric Toggles:
    - Sentiment Velocity: Tracking the 1–5 LLM-distilled sentiment score.
    - Attention Share: Plotting the % of discourse captured by a specific contestant.
    - Polarization Index: Visualizing the weekly standard deviation of sentiment to flag controversial peaks.
- Annotation Layer: Automated markers on the X-axis for "Villa Events" (e.g., Recoupling, Dumpings, Bombshell Entry) to correlate community sentiment shifts with show production milestones.
